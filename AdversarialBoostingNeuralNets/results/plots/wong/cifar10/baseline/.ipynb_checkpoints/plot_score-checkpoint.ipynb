```python
import matplotlib.pyplot as plt
```

```python
scores = [41.7, 42.8, 44.14, 44.34]
```

```python
wl_indices = list(range(1, len(scores) + 1))
```

```python
plt.subplots()
plt.plot(wl_indices, scores)
plt.xlabel("Weak learner")
plt.ylabel("Autoattack score")
plt.title("Autoattack score (independent model, eps = 0.031 = 8/255)")
plt.savefig("autoattack_cifar10.png", dpi=300)
```
